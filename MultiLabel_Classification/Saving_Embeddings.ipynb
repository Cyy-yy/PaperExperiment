```python
import torch
import numpy as np
import utils
from attention.model_add_cnn import StructuredSelfAttention
import data_got
```

```python
config = utils.read_config("config.yml")
```

```python
label_embed = torch.from_numpy(np.load("../Word2Vec/label_embeddings.npy")).float()
embed = torch.from_numpy(np.load("../Word2Vec/embeddings.npy")).float()

model = StructuredSelfAttention(
    batch_size=config.batch_size,
    lstm_hid_dim=config['lstm_hidden_dimension'],
    d_a=config["d_a"],
    n_classes=config.n_classes,
    label_embed=label_embed,
    embeddings=embed,
    embedding_size=config.embed_size,
    kernel_size=config.kernel_size,
    kernel_num=config.kernel_num
)
```

```python
model.load("checkpoints/checkpoint_LSAN_CNN.pth")
```

```python
model.embeddings.weight.requires_grad = False
embedding_trained = model.embeddings.weight.cpu().numpy()
```

```python
with open("../Word2Vec/vocab.txt", "r") as fp:
    vocab = fp.read().split()
```

```python
# Store trained label embeddings ["价格", "位置", "服务", "环境", "菜品"]

labels = ["价格", "位置", "服务", "环境", "菜品"]
label_embedding_trained = np.zeros([len(labels), 768], dtype=np.float32)
for i, label in enumerate(labels):
    idx = vocab.index(label)
    label_embedding_trained[i] = embedding_trained[idx]
```

```python
np.save("../Word2Vec/embeddings_trained", embedding_trained)
np.save("../Word2Vec/label_embeddings_trained", label_embedding_trained)
```
