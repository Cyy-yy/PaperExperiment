```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
train = pd.read_csv("./MultiLabel_Classification/data/train.csv").iloc[:, 2:]
dev = pd.read_csv("./MultiLabel_Classification/data/dev.csv").iloc[:, 2:]
test = pd.read_csv("./MultiLabel_Classification/data/test.csv").iloc[:, 2:]
train["tag"] = "train"
dev["tag"] = "dev"
test["tag"] = "test"
```

```python
df = pd.concat([train, dev, test])
df2plot = df.groupby("tag").sum().T.sort_values("train", ascending=False)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 4))
df2plot[["train", "dev", "test"]].plot(kind="bar", stacked=True, ax=ax)
ax.set_yticks([])
ax.set_xticklabels(df2plot.index, ha="right", rotation=45)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
fig.savefig("./MultiLabel_Classification/data/label_stat.jpg", bbox_inches="tight", dpi=300)
```
